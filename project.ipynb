**Part B**

```python
# This makes sure that every instance of null value (['quantity', 'price'] is filled with the median of the 'quantity' and 'price' columns of the dataset respectively)

transactions['quantity'] = transactions['quantity'].fillna(transactions['quantity'].median())
transactions['price'] = transactions['price'].fillna(transactions['price'].median())

# Mode finds the most frequent column attributes (in this case, the most common payment method) and returns a pandas series from most frequent to least frequent attributes
# Here, ...mode[0] is used because we select the most frequent attribute from the pandas series that we have got using the mode function
transactions['payment_method'] = transactions['payment_method'].fillna(transactions['payment_method'].mode()[0])
```

```python
transactions.isnull().sum()
```

```python
transactions['revenue'] = transactions['quantity']*transactions['price']
transactions['revenue'].head()
```

```python
# To ensure no data loss from the transactions table, we will apply left join

# First, lets left join transactions dataframe and customers dataframe on 'customer_id'
transactions_customers = pd.merge(transactions, customers, how='left', on='customer_id')

# Then we join the transactions_customers dataframe with the products dataframe on 'product_id
full_data = pd.merge(transactions_customers, products, how='left', on='product_id')
```

```python
full_data['profit_margin'] = ((full_data['price'] - full_data['cost_price'])/full_data['price'])*100
full_data['profit_margin'].head()
```

```python
latest_purchase = full_data.groupby('customer_id')['timestamp'].max().reset_index()
latest_purchase.columns = ['customer_id', 'last_purchase_date']

latest_purchase['recency'] = (pd.to_datetime('today') - latest_purchase['last_purchase_date']).dt.days

frequency = full_data.groupby('customer_id')['transaction_id'].count().reset_index()
frequency.columns = ['customer_id', 'frequency']

monetary = full_data.groupby('customer_id')['revenue'].sum().reset_index()
monetary.columns = ['customer_id', 'monetary']

rfm = pd.merge(latest_purchase[['customer_id', 'recency']], frequency[['customer_id', 'frequency']], on='customer_id')
rfm = pd.merge(rfm, monetary[['customer_id', 'monetary']], on='customer_id')
```

```python
rfm.head()
```

```python
# Assigning bins for respective columns by defining necessary ranges and it's respective labels

rfm['recency_bin'] = pd.cut(rfm['recency'], bins=[0, 30, 60, 90, 120, 9999], labels=['High', 'Medium', 'Low', 'Very Low', 'Very Low'], ordered = False)
rfm['frequency_bin'] = pd.cut(rfm['frequency'], bins=[0, 1, 3, 5, 10, 999], labels=['Low', 'Medium', 'High', 'Very High', 'Very High'], ordered = False)
rfm['monetary_bin'] = pd.cut(rfm['monetary'], bins=[0, 500, 1000, 5000, 10000, 99999], labels=['Low', 'Medium', 'High', 'Very High', 'Very High'], ordered = False)

def assign_segment(row):
    if row['recency_bin'] == 'High' and row['frequency_bin'] == 'High' and row['monetary_bin'] == 'High':
        return 'High Value'
    elif row['recency_bin'] == 'Low' and row['frequency_bin'] == 'Low' and row['monetary_bin'] == 'Low':
        return 'Low Value'
    elif row['recency_bin'] == 'Medium' and row['frequency_bin'] == 'Medium' and row['monetary_bin'] == 'Medium':
        return 'Medium Value'
    else:
        return 'Mixed'
    
    # Assigns customer value based on their recency, frequncy and monetary value
    
rfm['segment'] = rfm.apply(assign_segment, axis=1)
```

```python
# This function helps to find the suspicious transactions based on the conditions provided

def find_suspicious_transactions(full_data):

    condition_1 = (full_data['quantity'] > 100) & (full_data['price'] < 10)

    full_data['Hour_of_day'] = full_data['timestamp'].dt.hour
    
    purchase_count_per_hour = full_data.groupby(['customer_id', 'Hour_of_day']).size().reset_index(name='transaction_count')

    suspicious_customers = purchase_count_per_hour[purchase_count_per_hour['transaction_count'] > 3]

    full_data = full_data.merge(suspicious_customers[['customer_id', 'Hour_of_day']], 
                                on=['customer_id', 'Hour_of_day'], 
                                how='left', 
                                indicator=True)

    condition_2 = full_data['_merge'] == 'both'

    suspicious_mask = condition_1 | condition_2 # OR condition between condition 1 and condition 2

    suspicious_transactions = full_data[suspicious_mask]

    # .loc helps to select rows and columns by their labels

    suspicious_transactions.loc[:,'reason'] = "Quantity > 100 and Price < 10"
    suspicious_transactions.loc[condition_2, 'reason'] = "More than 3 purchases in a single hour"
    suspicious_transactions.loc[condition_1 & condition_2, 'reason'] = "Both conditions met"

    result = suspicious_transactions[['transaction_id', 'customer_name', 'reason']]

    return result
```

```python
# rolling helps to select a certain width/slice of the dataset and perform operations on it
# For Eg: We have calculated 7_day_moving_average by applying rolling on the daily_revenue and then averaging the sum (finding the mean)

full_data['date'] = full_data['timestamp'].dt.date
daily_revenue = full_data.groupby('date')['revenue'].sum().reset_index(name = 'daily_revenue')
daily_revenue['7_day_moving_avg'] = daily_revenue['daily_revenue'].rolling(window = 7, min_periods=1).mean()
print(daily_revenue.head())
```

```python
full_data['signup_date'] = pd.to_datetime(full_data['signup_date'])
first_transaction = full_data.groupby('customer_id')['timestamp'].min().reset_index() # min helps to select the earliest datetime
first_transaction.rename(columns={'timestamp': 'first_transaction_date'}, inplace=True)
first_transaction = pd.merge(first_transaction, full_data[['customer_id', 'signup_date']], on='customer_id', how='left')
first_transaction['days_since_signup'] = (first_transaction['first_transaction_date'] - first_transaction['signup_date']).dt.days

full_data = full_data.sort_values(by=['customer_id', 'timestamp'])
full_data['previous_purchase'] = full_data.groupby('customer_id')['timestamp'].shift(1)
full_data['days_between_purchases'] = (full_data['timestamp'] - full_data['previous_purchase']).dt.days

purchase_frequency = full_data.groupby('customer_id')['days_between_purchases'].mean().reset_index()
purchase_frequency.rename(columns={'days_between_purchases' : 'purchase_frequency'}, inplace = True)

category_diversity = full_data.groupby('customer_id')['category'].nunique().reset_index()
category_diversity.rename(columns={'category': 'category_diversity'}, inplace=True)

preferred_payment = full_data.groupby('customer_id')['payment_method'].agg(lambda x: x.mode()[0]).reset_index()
preferred_payment.rename(columns={'payment_method': 'preferred_payment'}, inplace=True)

customer_metrics = pd.merge(first_transaction[['customer_id', 'days_since_signup']], purchase_frequency[['customer_id', 'purchase_frequency']], on='customer_id', how='left')

customer_metrics = pd.merge(customer_metrics, category_diversity[['customer_id', 'category_diversity']], on='customer_id', how='left')

customer_metrics = pd.merge(customer_metrics, preferred_payment[['customer_id', 'preferred_payment']], on='customer_id', how='left')
```

```python
def customer_lifetime_value(transactions_df, customer_id, discount_rate = 0.1):
    customer_data = transactions_df[transactions_df['customer_id'] == customer_id]
    customer_data = customer_data.sort_values(by='timestamp')
    first_purchase = customer_data['timestamp'].min()
    customer_data['month_since_first_purchase'] = ((customer_data['timestamp'] - first_purchase).dt.days)//30
    customer_data['discounted_revenue'] = customer_data['revenue']*(1- discount_rate)**customer_data['Month']

    clv = customer_data['discounted_revenue'].sum()

    return clv

a1 = customer_lifetime_value(transactions, 1)
print(a1)
```
